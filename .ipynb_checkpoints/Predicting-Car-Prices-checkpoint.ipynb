# 数据清理
## 选取有用的列
有些列是我主观上觉得没有用的，比如“汽车制造商”和“汽油类型”。对于这些列，我选择直接删除了。```'normalized-losses', 'wheel-base', 'length', 'width',  'height', 'curb-weight', 
'num-of-cylinders', 'engine-size', 
'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg' ```                  
## 去除异常值
从上面的数据面可以看出来有些“奇怪”的值，比如‘？’当然可能还有一些用肉眼无法看出来。我选择用的方法是用unique函数整理出每一个列的唯一的字符串是什么，这样数据量会小很多，便于用肉眼正确观察。然后再删除异常的值。
## 数据类型转化
有一列 *'num-of-cylinders'（气缸数量）,eight, five, four, six, three, twelve, two.* 这看似str类型，实是数字型的，我们会先将它转为数字。然后将其他所有类型是数字类型的全都转为float。

````python
cols = ['normalized-losses', 'wheel-base', 'length', 'width',  'height', 'curb-weight', 
'num-of-cylinders', 'engine-size', 
'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg']

# 用有用列创建一个新的数据集
dataCatNew = dataCars[cols]
dataCatNew.head(3)

import numpy as np

# 筛选出每一列的唯一值
for col in cols:
    t = np.unique(dataCatNew[col])
    #print(col, ':',t, "\n")

# 只有一个异常值‘？’

# 输出没有删除前的数据集大小
print(dataCatNew.shape)

# 替换所有？为NaN
dataCatNew = dataCatNew.replace("?", np.nan)

# 删除所有异常值的行
dataCatNew = dataCatNew.dropna(axis=0)

# 输出删除后的数据集大小
print(dataCatNew.shape)

# 将str转为数字
#  'num-of-cylinders'（气缸数量）,eight, five, four, six, three, twelve, two.
dataCatNew["num-of-cylinders"] = dataCatNew["num-of-cylinders"].replace("eight", 8)
dataCatNew["num-of-cylinders"] = dataCatNew["num-of-cylinders"].replace("five", 5)
dataCatNew["num-of-cylinders"] = dataCatNew["num-of-cylinders"].replace("four", 4)
dataCatNew["num-of-cylinders"] = dataCatNew["num-of-cylinders"].replace("six", 6)
dataCatNew["num-of-cylinders"] = dataCatNew["num-of-cylinders"].replace("three", 3)
dataCatNew["num-of-cylinders"] = dataCatNew["num-of-cylinders"].replace("twelve", 12)
dataCatNew["num-of-cylinders"] = dataCatNew["num-of-cylinders"].replace("two", 1)

dataCatNew.head(3)
````

````python
# 所有数据转为浮点型
dataCatNew = dataCatNew.astype("float")
````
