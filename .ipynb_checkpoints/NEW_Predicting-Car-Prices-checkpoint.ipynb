# 数据清理
## 选取有用的列
有些列是我主观上觉得没有用的，比如“汽车制造商”和“汽油类型”。对于这些列，我选择直接删除了。```'normalized-losses', 'wheel-base', 'length', 'width',  'height', 'curb-weight', 
'num-of-cylinders', 'engine-size', 
'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg' ```                  
## 去除异常值
从上面的数据面可以看出来有些“奇怪”的值，比如‘？’当然可能还有一些用肉眼无法看出来。我选择用的方法是用unique函数整理出每一个列的唯一的字符串是什么，这样数据量会小很多，便于用肉眼正确观察。然后再删除异常的值。
## 数据类型转化
有一列 *'num-of-cylinders'（气缸数量）,eight, five, four, six, three, twelve, two.* 这看似str类型，实是数字型的，我们会先将它转为数字。然后将其他所有类型是数字类型的全都转为float。

````python
cols = ['normalized-losses', 'wheel-base', 'length', 'width',  'height', 'curb-weight', 
'num-of-cylinders', 'engine-size', 
'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg', 'price']

# 用有用列创建一个新的数据集
dataCatNew = dataCars[cols]
dataCatNew.head(3)

import numpy as np

# 筛选出每一列的唯一值
for col in cols:
    t = np.unique(dataCatNew[col])
    #print(col, ':',t, "\n")

# 只有一个异常值‘？’

# 输出没有删除前的数据集大小
print(dataCatNew.shape)

# 替换所有？为NaN
dataCatNew = dataCatNew.replace("?", np.nan)

# 删除所有异常值的行
dataCatNew = dataCatNew.dropna(axis=0)

# 输出删除后的数据集大小
print(dataCatNew.shape)

# 将str转为数字
#  'num-of-cylinders'（气缸数量）,eight, five, four, six, three, twelve, two.
dataCatNew["num-of-cylinders"] = dataCatNew["num-of-cylinders"].replace("eight", 8)
dataCatNew["num-of-cylinders"] = dataCatNew["num-of-cylinders"].replace("five", 5)
dataCatNew["num-of-cylinders"] = dataCatNew["num-of-cylinders"].replace("four", 4)
dataCatNew["num-of-cylinders"] = dataCatNew["num-of-cylinders"].replace("six", 6)
dataCatNew["num-of-cylinders"] = dataCatNew["num-of-cylinders"].replace("three", 3)
dataCatNew["num-of-cylinders"] = dataCatNew["num-of-cylinders"].replace("twelve", 12)
dataCatNew["num-of-cylinders"] = dataCatNew["num-of-cylinders"].replace("two", 1)

dataCatNew.head(3)
````

````python
# 所有数据转为浮点型
dataCatNew = dataCatNew.astype("float")
````

# 对单独每一个特征探究
## 检测方法
RMSE
## 思路
我准备做一个表格，行是k值从1至100，列是每一个特征，每一小格填写该k值该特征的RMSE。然后画图。

#### 增加一个新的特征面积和体积
发现特征中有长宽高这三个变量，马上联想到面积和体积是两个新的特征

````python
dataCatNew["mianji"] = dataCatNew["length"]*dataCatNew["width"]
dataCatNew["tiji"] = dataCatNew["mianji"]*dataCatNew["height"]
dataCatNew.head(3)
````

````python
from sklearn.neighbors import KNeighborsRegressor
from sklearn.metrics import mean_squared_error

# KNN模型，输入不同的特征，输出不同的RMSE
# train_col:  特征列
# kval : k值
# target_col:  目标列 (price)
# data: 数据集
# return RMSE
def knn_model(train_col, kval, target_col, data):
    # 获取数据集的大小
    data_shape = data.shape
    
    # 乱序排序每一行数据
    np.random.seed(1)
    shuffled_index = np.random.permutation(data.index)
    data = data.reindex(shuffled_index)
    
    # 80%训练数据集 20%测试数据集
    train_data = data[1 : int(data_shape[0]*0.8)]
    test_data = data[int(data_shape[0]*0.8) : ]
    
    # 创建模型，并且做出预测列p
    knn = KNeighborsRegressor(n_neighbors=kval)
    knn.fit(train_data[train_col], train_data[target_col])
    p = knn.predict(test_data[train_col])
    
    # 计算RMSE
    rmse= (mean_squared_error(p, test_data[target_col]))**(1/2)
    return rmse

# 某一特征的某一k值的RMSE值构成的字典
col_k_rmse = {}

# 特征列
cols = ['normalized-losses', 'wheel-base', 'length', 'width',  'height', 'curb-weight', 
'num-of-cylinders', 'engine-size', 
'bore', 'stroke', 'compression-rate', 'horsepower', 'peak-rpm', 'city-mpg', 'highway-mpg',"mianji","tiji"]

# 先做好一个小字典，然后将小字典插入大字典
for col in cols:
    dic = {}
    for k in range(1,20):
        rmse=knn_model([col], k , ["price"], dataCatNew)
        dic[k]=rmse
    col_k_rmse[col]=dic
data_col_k_rmse = pd.DataFrame(col_k_rmse) 
data_col_k_rmse.head(3)
````

````python
import matplotlib.pyplot as plt
fig, ax1 = plt.subplots(figsize=(15,15))

# 画出一个折线图，关于不同特征的k和RMSE的折线图
# 由于颜色不够用所以增加线的样式
for col in cols[0:6]:
    ax1.plot(range(1,20), data_col_k_rmse[col],"--")
for col in cols[6:11]:
    ax1.plot(range(1,20), data_col_k_rmse[col])
for col in cols[11:]:
    ax1.plot(range(1,20), data_col_k_rmse[col], 'o-')
    
plt.legend(loc="upper right")
plt.xlabel("k")
plt.ylabel("RMSE")
plt.title(u"不同列的不同k的RMSE值")
````

### 小结
单独一个特征中'highway-mpg'特征在k约=6时表现是最好的。新增加的特征中“mianji”特征在k约=3时表现最好。以上表现最好指的是RMSE值很低。

#### 计算每一个特征的平均RMSE
- 1、计算平均RMSE
- 2、找到整体RMSE最小的几个特证
- 3、画出平均RMSE的条形图

````python
dic = {}
for col in cols:
    dic[col] = data_col_k_rmse[col].mean()
    
# avg_RMSE是每一个特征的k值为1-10的平均RMSE
avg_RMSE=pd.Series(dic)

# 肉眼观察2500以下是最好的RMSE
avg_RMSE[avg_RMSE<2500]
````

## 小结
通过上面的工作我们可以确定出

```city-mpg       2307.075632
highway-mpg    2069.393678
mianji         2489.637840```
是比较好的特征。

# 对多个特征探究
我们将多个特征组合在一起，然后通过枚举K的值，来看看当k取值多少时可以取到最小的RMSE。

````python
# 特征列
cols = ["city-mpg", "highway-mpg", "mianji"]

for k in range(1,5):
    RMSE = knn_model(cols, k, "price", dataCatNew)
    print("k=", k, ", RMSE=", RMSE,"\n")
````

# 总结
对于knn模型而言我们，我们k的值取**3**，特征使用 "city-mpg", "highway-mpg", "mianji",能达到比较好的效果

````python
# return 预测值和实际值的误差
def knn_model(train_col, kval, target_col, data):
    # 获取数据集的大小
    data_shape = data.shape
    
    # 乱序排序每一行数据
    np.random.seed(1)
    shuffled_index = np.random.permutation(data.index)
    data = data.reindex(shuffled_index)
    
    # 80%训练数据集 20%测试数据集
    train_data = data[1 : int(data_shape[0]*0.8)]
    test_data = data[int(data_shape[0]*0.8) : ]
    
    # 创建模型，并且做出预测列p
    knn = KNeighborsRegressor(n_neighbors=kval)
    knn.fit(train_data[train_col], train_data[target_col])
    p = knn.predict(test_data[train_col])
    return p-test_data["price"]

wucha=knn_model(["city-mpg", "highway-mpg", "mianji"], 3, "price", dataCatNew)
print(wucha, "\n平均误差=",wucha.mean())
````

总体来看是一个不错的模型
